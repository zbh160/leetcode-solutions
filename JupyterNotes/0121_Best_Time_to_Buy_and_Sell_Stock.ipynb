# LC121 - Best Time to Buy and Sell Stock
**Pattern:** Greedy • Running Minimum


## Key Observation
Only the lowest price seen so far matters.


```python
prices=[7,1,5,3,6,4]
min_price=float('inf')
max_profit=0
for p in prices:
    min_price=min(min_price,p)
    profit=p-min_price
    max_profit=max(max_profit,profit)
    print(p,min_price,profit,max_profit)
```

```python
class Solution(object):
    def maxProfit(self, prices):
        max_profit=0
        min_price=float('inf')
        for price in prices:
            min_price=min(min_price,price)
            max_profit=max(max_profit,price-min_price)
        return max_profit
```

## Complexity
- Time: **O(n)**
- Space: **O(1)**
