# LC238 - Product of Array Except Self

**Pattern:** Prefix & Suffix


## Key Observation

For each index: `answer[i] = prefix product × suffix product`.


## Step 1: Build Prefix Products

Store the running product of all values before each index.


```python
nums = [1, 2, 3, 4]
result = [1] * len(nums)
prefix = 1
for i in range(len(nums)):
    result[i] = prefix
    print(f'i={i}, prefix={prefix}, result={result}')
    prefix *= nums[i]
print('Prefix result:', result)
```

## Step 2: Multiply Suffix Products

Traverse from right to left and multiply the running suffix into `result[i]`.


```python
suffix = 1
for i in range(len(nums)-1, -1, -1):
    result[i] *= suffix
    print(f'i={i}, suffix={suffix}, result={result}')
    suffix *= nums[i]
print('Final result:', result)
```

## Accepted Solution


```python
class Solution(object):
    def productExceptSelf(self, nums):
        result = [1] * len(nums)
        prefix = 1
        for i in range(len(nums)):
            result[i] = prefix
            prefix *= nums[i]
        suffix = 1
        for i in range(len(nums)-1, -1, -1):
            result[i] *= suffix
            suffix *= nums[i]
        return result
```

## Complexity

- Time: **O(n)**
- Space: **O(1)** excluding the output array
