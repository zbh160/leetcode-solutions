# Two Sum

## Problem
Find the indices of two numbers whose sum equals the target.

## Brute Force

Try every pair of numbers.

- Time: **O(n²)**
- Space: **O(1)**

## Key Observation

Instead of searching twice, remember the numbers you've already seen using a hash map.

```python
nums = [2, 7, 11, 15]
target = 9
hashmap = {}

for i, num in enumerate(nums):
    diff = target - num
    print(f"Index={i}, Num={num}, Need={diff}, HashMap={hashmap}")
    if diff in hashmap:
        print("Found:", [hashmap[diff], i])
        break
    hashmap[num] = i
```

## Accepted Solution

```python
class Solution(object):
    def twoSum(self, nums, target):
        hashmap = {}
        for i, num in enumerate(nums):
            diff = target - num
            if diff in hashmap:
                return [hashmap[diff], i]
            hashmap[num] = i
        return []
```

## Complexity

- Time: **O(n)**
- Space: **O(n)**
