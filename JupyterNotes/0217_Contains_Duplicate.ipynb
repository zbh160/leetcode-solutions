# LC217 - Contains Duplicate
**Pattern:** Hash Set


## Key Observation
We only need to know whether a number has appeared before.


```python
nums=[1,2,3,1]
seen=set()
for n in nums:
    print('Current:',n,' Seen:',seen)
    if n in seen:
        print('Duplicate Found:',n)
        break
    seen.add(n)
```

## Accepted Solution


```python
class Solution(object):
    def containsDuplicate(self, nums):
        num_show=set()
        for num in nums:
            if num in num_show:
                return True
            num_show.add(num)
        return False
```

## Complexity
- Time: **O(n)**
- Space: **O(n)**
